## Pandas

Pandas is a Python library used for working with data—especially tables, spreadsheets, and time-series data.
## What Pandas is mainly used for

1. Reading data (CSV, Excel, JSON, SQL, etc.)
2. Cleaning messy data (missing values, duplicates)
3. Analyzing data (filters, groups, summaries)
4. Manipulating data (add/remove columns, sort, merge)
5. Basic data exploration before ML or visualization

### Why Pandas is so popular

1. Very easy to use
2. Fast and efficient
3. Works great with NumPy, Matplotlib, Seaborn
4. Widely used in Data Science, ML, AI, Finance, DevOps analytics
