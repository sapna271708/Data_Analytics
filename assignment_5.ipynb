## Sales Data Analysis Dashboard

<img src="./Screenshot 2026-03-07 111456.png">

### The dashboard is designed in Microsoft Power BI to analyze sales data and help users understand:

- Which products sell the most
- Which regions generate the most sales
- The total overall sales

### Slicers (Filters)

On the left and right side at the top, there are two Slicers.

#### Product Slicer

This filter allows the user to select a specific product, such as:

- Laptop
- Mobile
- Tablet
- Headphones

If we choose a product, all charts in the dashboard will update and show sales related only to that product.

#### Region Slicer

This filter allows the user to select a specific region, such as:

- North
- South
- East
- West

Selecting a region will update all visuals to display sales data for that region only.

These slicers make the dashboard interactive and dynamic.

### Card Visual (Total Sales)

In the center at the top, there is a Card Visual showing “6M Sum of Sales.”

This means:

Total Sales = 6 Million

This card provides a quick summary of overall sales in the dataset.

### Bar Chart – Sales by Product

The chart on the left side is a Bar Chart titled “Sum of Sales by Product.”

This chart shows how much revenue each product generates.

From the chart we can observe:

- Laptop has the highest sales
- Mobile has the second highest sales
- Tablet has moderate sales
- Headphones have the lowest sales

### Pie Chart – Sales by Region

On the right side, there is a Pie Chart titled “Sum of Sales by Region.”

This chart shows how total sales are distributed across different regions.

From the pie chart:

- North Region ≈ 2M sales
- South Region ≈ 2M sales
- East Region ≈ 1M sales
- West Region ≈ 1M sales

#### This dashboard helps us understand three main things:

- Total sales performance (6 Million).

- Product performance (Laptop is the top-selling product).

- Regional performance (North and South generate the highest sales).

Because the dashboard includes filters, users can interact with the data and analyze specific products or regions easily.
