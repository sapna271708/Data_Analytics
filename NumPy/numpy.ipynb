### Array Attributes

```python
# creating a 1D array of numbers from 0 to 59
arr=np.arange(60) # arange is used to create an array of numbers from 0 to n-1
```

```python
# reshaping the array to 12 rows and 5 columns
arr = arr.reshape(12,5)
```

```python
# displaying the array
arr
```

```python
print("shape of the array is ", arr.shape) # shape is used to get the number of rows and columns in the array
print("data type of the array is ", arr.dtype) # dtype is used to get the data type of the array
print("number of dimensions of the array is ", arr.ndim) # ndim is used to get the number of dimensions of the array
print("size of the array is ", arr.size) # size is used to get the total number of elements in the array
```

```python
arr[1] # indexing the array to get the second row of the array
```

```python
arr[1][3] # indexing the array to get the element in the second row and fourth column of the array
```

```python
arr[-11][-1] # indexing the array to get the element in the second row and fourth column of the array using negative indexing
```

```python
arr[11,4] # indexing the array to get the element in the last row and last column of the array using negative indexing
```

```python
arr[-1,-1]=20 #
```

```python
arr # displaying the array after changing the last element to 20
```

```python
arr.T # transposing the array to get the columns as rows and rows as columns
```

### convertion of arrays 

```python
arr=arr.flatten() # flattening the array to get a 1D array from a 2D array
```

```python
arr # displaying the flattened array
```

```python
print("shape of the array is ", arr.shape) # shape is used to get the number of rows and columns in the array
print("data type of the array is ", arr.dtype) # dtype is used to get the data type of the array
print("number of dimensions of the array is ", arr.ndim) # ndim is used to get the number of dimensions of the array
print("size of the array is ", arr.size) # size is used to get the total number of elements in the array
```

```python
arr=np.arange(225).reshape(5,15,3) # creating a 3D array of numbers from 0 to 224 and reshaping it to 5 layers, 15 rows and 3 columns
```

```python
arr # displaying the flattened array
```

```python
arr[0] # indexing the array to get the first layer of the 3D array
```

```python
arr[0,0,1] # indexing the array to get the element in the first layer, first row and second column of the 3D array
```

```python
arr[0:4,0:3,0:2] # slicing the array to get the first 4 layers, first 3 rows and first 2 columns of the 3D array
```

```python
arr[0:4,0,0] # slicing the array to get the first 4 layers, first row and first column of the 3D array
```

```python
arr[0:5,0:5,0:3] # slicing the array to get the first 5 layers, first 5 rows and first 3 columns of the 3D array
```

```python
arr[0:4][0][0] # slicing the array to get the first 4 layers, first row and first column of the 3D array using multiple indexing
```
