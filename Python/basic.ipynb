### Data types

Python has following datatypes:
1. Numeric Data Type
 - int : whole numbers , integers
 - float : point values , Decimal numbers
 - complex : Real + imaginary part
2. Text Data Type
 - string : Used for text / characters
3. Boolean Data Type
 - True
 - False
4. Sequence Data Types
 - list : Ordered, changeable
 - tuple : Ordered, not changeable
5. Set Data Type
 - set : Unordered, unique values only
6. Mapping Data Type
 - dict : dictonary (key-pair)
7. None Data Type
 - none : no value

```python
x = 10        # int
y = 3.5       # float
a = 5 + 2j    # complex
z = "Python"  # string
s = [1, 2, 3]  # list
t = (4, 5, 6)  # tuple
d = {"name": "Alice", "age": 30}  # dictionary
f = {1, 2, 3}  # set
n = None       # NoneType
is_ok = True  # boolean
```
