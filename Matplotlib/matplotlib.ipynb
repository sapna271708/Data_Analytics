```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a=np.array([1,2,3,4,5])
s=np.array([2,4,6,8,10])
```

```python
a
```

```python
s
```

```python
plt.plot(a,s)
plt.show()
```

```python
plt.scatter(a,s)
```

```python
plt.plot(a, marker='*', linestyle='--', color='g')
```

```python
plt.plot(s, marker='o', linestyle='--', color='b')
```

```python
plt.plot(s, marker='o', linestyle='--', color='b',ms=10)
```

```python
plt.plot(s, marker='o', linestyle='--', color='b',mec='r',ms=10)
```

```python
plt.plot(s, marker='o', linestyle='--', color='b', label='s values')
```

```python
plt.plot(s, marker='o', linestyle='--', color='b', mfc='g', ms=10,)
```

```python
plt.plot(a,linewidth=5)
```

```python
plt.plot(a)
plt.plot(s,linewidth=5)
```
