## 1. Explain: Data Cleaning & Preprocessing

• Load the dataset and check for missing values.

• Fill missing Age values with median or mean.

• Handle missing Embarked values by filling with the most common port.

• Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the dataset 

```python
# Load the dataset
data = pd.read_csv('train.csv')
```

```python
# display the dataset
data.head()
```

### Check the Missing Values

```python
# Check for missing values
data.isnull().sum()
```

### Fill missing Age values with median or mean.


```python
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
data.head()
```

### Handle missing Embarked values by filling with the most common port.

```python
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

### Drop unnecessary columns like Name, Ticket, Cabin for visualization purpose

```python
data.drop(columns=['Cabin','Name','Ticket'], inplace=True)
```

```python
data.head()
```

## 2. Univariate Analysis
* Plot the distribution of Age using a histogram and a KDE plot.
* Plot the count of passengers by Sex using a bar chart.
* Show the distribution of Fare using a boxplot.

### 1. Plot the distribution of Age using a Histogram and KDE plot

```python

plt.figure()
sns.histplot(data['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

### Plot the count of passengers by Sex using a bar chart.

```python
sns.countplot(x='Sex', data=data)
plt.title("Count of passengers by Sex")
plt.show()
```

### Show the distribution of Fare using a boxplot.

```python
sns.boxplot(x='Fare', data=data)
plt.title("Boxplot of Fare")
plt.show()
```

```python
sns.boxplot(y='Fare', data=data)
plt.title("Boxplot of Fare")
plt.show()
```

## 3. Bivariate Analysis
* Compare survival rate across Sex using a countplot.
* Compare Pclass vs Fare using a boxplot.
* Visualize Age vs Fare using a scatter plot, color-coded by Survived

### Compare survival rate across Sex using a countplot

```python
data.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival Count by Sex")
plt.show()
```

### Compare Pclass vs Fare using a boxplot.

```python
sns.boxenplot(x='Pclass', y='Fare', data=data)
plt.title("Fare Distribution by Pclass")
plt.show()
```

### Visualize Age vs Fare using a scatter plot, color-coded by Survived

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=data)
plt.title("Age vs Fare colored by Survival")
plt.show()
```
